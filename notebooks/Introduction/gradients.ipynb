```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
x = jnp.linspace(-3, 3, 100)
```

```python
y = jnp.sin(x)
```

```python
plt.plot(x, y, label='y=sin(x)')
plt.legend()
```

```python
dydx = jax.vmap(jax.grad(jnp.sin))
```

```python
dydx(jnp.array([0.0, 0.1]))
```

```python
plt.plot(x, y, label='y=sin(x)')
plt.plot(x, dydx(x), label='y=dsin(x)/dx')

plt.legend()
```
